```python
import pandas as pd
import glob
import os
```

```python
ratings = pd.read_csv('/content/sample_data/BX-Book-Ratings.csv', error_bad_lines=False, delimiter=';', encoding = 'ISO-8859-1')
```

```python
books = pd.read_csv('/content/sample_data/BX-Books.csv', error_bad_lines=False, delimiter=';', encoding = 'ISO-8859-1')
```

```python
ratings
```

```python
data = pd.merge(books, ratings, on='ISBN', how='inner')
```

```python
data = data.iloc[:-990000 , :]
```

```python
data
```

```python
data=data.drop(columns=['ISBN','Publisher','Image-URL-S','Image-URL-L'])
```

```python
data
```

```python
userRatings = data.pivot_table(index=['User-ID'],columns=['Book-Title'],values='Book-Rating')
userRatings.head()
```

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=20)
corrMatrix.head()
```

```python
myRatings = data.loc[data['User-ID']==].dropna()
myRatings
```

```python
myRatings = myRatings.set_index("Book-Title")
myRatings.index[1]
```

```python
similar_candidates = pd.Series()
for i in range(0, len(myRatings.index)):
    print("Adding similar book for " + myRatings.index[i] + "...")
    # retrieve similar movies to this one that I rated
    similar_movies = corrMatrix[myRatings.index[i]].dropna()
    # scale its similarity by how well I rated this movie
    similar_movies = similar_movies.map(lambda x: x * myRatings.iloc[i]['Book-Rating'])
    # add the score to the list of similar candidates
    similar_candidates = similar_candidates.append(similar_movies)
    
print("Sort recommendations...")
similar_candidates.sort_values(inplace = True, ascending = False)
similar_candidates.head(10)
```

```python
similar_candidates = similar_candidates.groupby(similar_candidates.index).sum()
similar_candidates = similar_candidates.drop(similar_candidates.loc[similar_candidates.index.isin(myRatings.index)].index)
```

```python
similar_candidates.sort_values(inplace = True, ascending = False)
similar_candidates.head(10)
```

```python
compression_opts = dict(method='zip', archive_name='outCorr.csv')  
corrMatrix.to_csv('out.zip', index=True, compression=compression_opts)  
```
